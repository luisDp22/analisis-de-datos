```python
import matplotlib.pyplot as plt
from sklearn import linear_model
import numpy as np
import seaborn as ans
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
df
```

```python
df.plot.scatter(x="metro", y="precio")
plt.show()
```

```python
df.columns
df.info
df.describe
```

```python
regresion = linear_model.LinearRegression()
```

```python
lcd = df["metro"].values.reshape((-1,1))
```

```python
modelo = regresion.fit(lcd, df["precio"])
print("Interseccion (b)", modelo.intercept_)
print("Pendiente (m)", modelo.coef_)
```

```python
entrada = [[7.5],[10.0],[12.5],[15.0],[22.5],[25.0]]
predicciones = modelo.predict(entrada)
print(predicciones)
```

```python
df.plot.scatter(x="metro", y="precio", label='Datos Originales')
plt.scatter(entrada, predicciones, color='red')
plt.plot(entrada, predicciones, color='black', label='Linea de regresión')
plt.xlabel('metro')
plt.ylabel('precio')
plt.legend()
plt.show()
```
