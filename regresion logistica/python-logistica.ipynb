```python
import matplotlib.pyplot as plt
from sklearn import linear_model
import numpy as np
import seaborn as ans
import pandas as pd
```

```python
df = pd.read_csv("framingham.csv")
df
```

```python
df.plot.scatter(x="sysBP", y="diaBP")
plt.show()
```

```python
df.columns
df.info
df.describe
```

```python
regresion = linear_model.LinearRegression()
```

```python
lcd = df["sysBP"].values.reshape((-1,1))
```

```python
modelo = regresion.fit(lcd, df["diaBP"])
print("Interseccion (b)", modelo.intercept_)
print("Pendiente (m)", modelo.coef_)
```

```python
entrada = [[90],[175],[200],[250],[280],[300]]
predicciones = modelo.predict(entrada)
print(predicciones)
```

```python
df.plot.scatter(x="sysBP", y="diaBP", label='Datos Originales')
plt.scatter(entrada, predicciones, color='red')
plt.plot(entrada, predicciones, color='black', label='Linea de regresiÃ³n')
plt.xlabel('sysBP')
plt.ylabel('diaBP')
plt.legend()
plt.show()
```
